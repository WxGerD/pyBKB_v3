```python
# -*- coding: utf-8 -*-
"""
Map tile acquisition
--------------------

Demonstrates cartopy's ability to draw map tiles which are downloaded on
demand from the Stamen tile server. Internally these tiles are then combined
into a single image and displayed in the cartopy GeoAxes.

"""
import matplotlib.pyplot as plt
from matplotlib.transforms import offset_copy

import cartopy.crs as ccrs
import cartopy.io.img_tiles as cimgt
```

```python
#stamen_styles = ['terrain-background', 'terrain', 'toner', 'watercolor']
#google_style = ['street', 'satellite', 'terrain', 'only_streets']
```

```python
## Create a Stamen Terrain instance.

tile = cimgt.StamenTerrain()

#tile = cimgt.GoogleTiles()
#tile = cimgt.OSM()
#tile = cimgt.QuadtreeTiles()


###tile = cimgt.MapQuestOSM() ##!!Doesn't work
```

```python
fig = plt.figure(figsize=(16,8))

# Create a GeoAxes in the tile's projection.
ax1 = fig.add_subplot(1, 2, 1, projection=tile.crs)
ax2 = fig.add_subplot(1, 2, 2, projection=tile.crs)

# Limit the extent of the map to a small longitude/latitude range.
lat = 40.77069
lon = -111.96503

### Zoomed in close
size = .3

ax1.set_extent([lon-size, lon+size, lat-size, lat+size])

# Add the Stamen data at zoom level 8.
ax1.add_image(tile, 10)
# Add a marker for the Eyjafjallajökull volcano.
ax1.plot(lon, lat, marker='o', color='red', markersize=12,
         alpha=0.7, transform=ccrs.Geodetic())


### Zoomed out level
size = 5
ax2.set_extent([lon-size, lon+size, lat-size, lat+size])

# Add the Stamen data at zoom level 8.
ax2.add_image(tile, 6)
# Add a marker for the Eyjafjallajökull volcano.
ax2.plot(lon, lat, marker='o', color='red', markersize=12,
         alpha=0.7, transform=ccrs.Geodetic())
```
