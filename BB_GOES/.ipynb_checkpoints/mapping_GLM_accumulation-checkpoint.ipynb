# Map the GLM flash locations 

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

import sys
sys.path.append('/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v3')
from BB_GOES.get_GOES import get_GOES_nearesttime
from BB_GOES.get_GLM import get_GLM_file_nearesttime, accumulate_GLM_FAST, filter_by_HRRR
from BB_maps.my_basemap import draw_centermap, draw_HRRR_map
```

```python
# Make Map
m = draw_centermap('WBB', size=[1,1])
```

```python
DATE = datetime(2019, 8, 30, 4, 30)
SAVEDIR = '/uufs/chpc.utah.edu/common/home/u0553130/public_html/Events/20190830_Lightning/'
```

```python
GLM_files = get_GLM_file_nearesttime(DATE, window=60)
```

```python
GLM_flash = accumulate_GLM_FAST(GLM_files, data_type='flash')
```

```python
GLM_events = accumulate_GLM_FAST(GLM_files, data_type='event')
```

```python
GLM
```

```python
plt.figure(figsize=[7,7])
m.drawstates(linewidth=2)
m.drawcounties(linewidth=1)
m.arcgisimage(service='World_Shaded_Relief', xpixels=1000)

m.scatter(GLM_flash['longitude'], GLM_flash['latitude'],
          edgecolor='k',
          latlon=True)

plt.title('GLM-East Flashes', fontweight='semibold', loc='left')
plt.title('Beginning: %s\n   Ending: %s' % (GLM_flash['DATETIME'][0].strftime('%d %b %Y %H:%M UTC'),
                                            GLM_flash['DATETIME'][1].strftime('%d %b %Y %H:%M UTC')), loc='right')

plt.savefig(SAVEDIR+'flashes_G16')
```

```python
plt.figure(figsize=[7,7])
m.drawstates(linewidth=2)
m.drawcounties(linewidth=1)
m.arcgisimage(service='World_Shaded_Relief', xpixels=1000)

m.scatter(GLM_events['longitude'], GLM_events['latitude'],
          edgecolor='k',
          latlon=True)

plt.title('GLM-East Events', fontweight='semibold', loc='left')
plt.title('Beginning: %s\n   Ending: %s' % (GLM_events['DATETIME'][0].strftime('%d %b %Y %H:%M UTC'),
                                            GLM_events['DATETIME'][1].strftime('%d %b %Y %H:%M UTC')), loc='right')

plt.savefig(SAVEDIR+'events_G16')
```
