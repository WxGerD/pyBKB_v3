# Compute HRRR-GLM Lightning False Alarm
How much area of forecasted lightning had not GLM flashes occur inside the area of predicted lightning?

Use the MapBox GeoJSON-Area code for computing the area of each contour.
- This method accounts for the curvature of the earth, whereas the standard Green's Theorem method is for a plane. 
- You can input the latitude-longitude points directly, rather than converting the vertices to a cartesian grid.

---

Steps:
1. Total number of flashes in HRRR (or other) domain.
2. Create contours. Vertices are latitude/longitude points.
3. For each contour
    1. What is the area? += Fxx_Area_Total (sum the total area of flash area)
    1. Does a GLM flash exist in it? If yes, += Fxx_Area_w_Flash (sum the total area with flashes occuring)
    1. How many flashes exist in it? += Fxx_inside
    
    
    
---

**HRRR Hit Rate**: Percentage of GLM flashes that occured inside a forecasted LTNG contour.  
**HRRR False Alarm Rate**: Percentage of total forecasted area with no GLM flash.

```python
# =============================================================================
# =============================================================================
print('Make domain paths...')
## Create Path boundaries of HRRR domain and subdomains of interest:
# HRRR: All points counter-clockwise around the model domain.
# West, Central, East: A 16 degree wide and 26 degree tall boundary region.
PATH_points = {
    'HRRR':
        {'lon': np.concatenate([Hlon[0], Hlon[:,-1], Hlon[-1][::-1], Hlon[:,0][::-1]]),
         'lat': np.concatenate([Hlat[0], Hlat[:,-1], Hlat[-1][::-1], Hlat[:,0][::-1]])},
    'West':{
        'lon':[-120, -104, -104, -120, -120],
        'lat':[24.4, 24.4, 50.2, 50.2, 24.2]},
    'Central':{
        'lon':[-104, -88, -88, -104, -104],
        'lat':[24.4, 24.4, 50.2, 50.2, 24.2]},
    'East':{
        'lon':[-88, -72, -72, -88, -88],
        'lat':[24.4, 24.4, 50.2, 50.2, 24.2]},
    'Utah':{
        'lon':[-114.041664, -111.047526, -111.045645,  -109.051460, -109.048632, -114.051534, -114.041664],
        'lat':[41.993580, 42.002846, 40.998538, 40.998403, 36.998310, 37.000574, 41.993580]}
}

## Combine lat/lon as vertice pair as a tuple. i.e. (lon, lat).
PATH_verts = {}
for i in PATH_points.keys():
    PATH_verts[i] = np.array([(PATH_points[i]['lon'][j], PATH_points[i]['lat'][j]) for j in range(len(PATH_points[i]['lon']))])

## Generate Path objects from the vertices.
PATHS = {}
for i in PATH_verts.keys():
    PATHS[i] = Path(PATH_verts[i])

print('                    ...Done')
```

```python
## Filter GLM obsrvations within each domain
print("Total GLM observations:", len(glm['latitude']))

# Generate a lat/lon tuple for each point
GLM_latlon_pair = [(glm['longitude'][i], glm['latitude'][i]) for i in range(len(glm['latitude']))]
```

```python
# Return a None if the legnth of GLM_latlon_pairs is zero, because there
# are no flashes.    
if len(GLM_latlon_pair) == 0:
    print('!!! Warning !!! DATE %s had no lightning data' % DATE)

## Determine which GLM observation points, from a lat/lon tuple, are in
# each boundary Path.
# !! Refer to the Paths generated at the begining of the script!!

# Using each Path in PATHS, determine which GLM points fall inside the Path
inside_path = {}
for i in PATHS.keys(): 
    inside_path[i] = PATHS[i].contains_points(GLM_latlon_pair)

## Filter the GLM data keys by the bounding Path (points inside Path):
filtered_glm = {}
for i in inside_path.keys():
    filtered_glm[i] = {
        'latitude': glm['latitude'][inside_path[i]],
        'longitude': glm['longitude'][inside_path[i]],
        'energy': glm['energy'][inside_path[i]],
        'area': glm['area'][inside_path[i]]
    }
```

```python
## Create Contour Collections Objects -------------------------------------
#--------------------------------------------------------------------------
# We can matplotlib.pyplot.contour() to generate Paths for each contour
# area for every forecast lead time requested. These paths are used to
# determine if a GLM flash occurred within the HRRR forecasted lightning
# threat area.
# NOTE: We must generate the collections with a cylindrical map projection,
#       else contour returns funny map coordinates for the vertices of the
#       path.
# NOTE: Use H.data to access the unmasked array for correct contouring.
# NOTE: Set levels=[0] to draw contours where lightning threat >0.

paths_glm = filtered_glm

domain_CONTOURS = {}
for i in filtered_glm.keys():
    domain_CONTOURS[i] = []
    if i == 'HRRR':
        for j, H in enumerate(HH):
            c = plt.contour(Hlon, Hlat, H.data, levels=[0]) # Just make contours from un-projected grid.
            domain_CONTOURS[i].append(c)
        plt.show()
    else:
        for j, H in enumerate(HH):
            # Get a copy of the gridded forecast data
            subDOMAIN = H.data.copy()
            # Which model grid points are inside the path?
            in_path = PATHS[i].contains_points(list(zip(Hlon.flatten(), Hlat.flatten())))
            # Which model grid points are not inside the path?
            not_in_path = np.invert(in_path)
            # Reshape points not in path boolean to shape of HRRR grid
            not_in_path = not_in_path.reshape(np.shape(H))
            # Set all points not in path to zero
            subDOMAIN[not_in_path] = 0
            # Make contours based on LTNG data inside the path.
            c = plt.contour(Hlon, Hlat, subDOMAIN, levels=[0]) # Just make contours from un-projected grid.
            domain_CONTOURS[i].append(c)
        plt.show()
```

```python
paths_glm = filtered_glm
COLLECTIONS = CONTOURS

## GLM Path Dictionary, Contour Collection, Lead Time, Contour Level Index
# Length of inputs should equal the number of contour collections which 
# should be the number of forecast lead times, i.e. len(fxx).
inputs = [[paths_glm, COL, level] for i, COL in enumerate(COLLECTIONS)]
```

```python
#=============================================================
# There is a CONTOUR for each FXX. Specify one to look at.
this_FXX = 5
#=============================================================

level = 0 # generally, this is zero, unless you make more contour levels.
COLLECTION = COLLECTIONS[this_FXX]
```

```python
fxx_stats = {}
for i in paths_glm.keys():
    # Total Number of GLM Flashes
    num_flashes = len(paths_glm[i]['latitude'])
    #
    # Initialize a dictionary to store statistics for this forecast grid.
    #     flash inside    : An array indicating which GLM flashes occur 
    #                       inside a HRRR forecasted area. Initialized as zeros
    #                       with length of number of GLM flashes.
    #     area total      : Total area of HRRR forecasted LTNG (accumulates over loop)
    #     area with flash : Total area of HRRR LTNG contours that contain a GLM flash.
    #                       (accumulates over loop).    
    fxx_stats[i] = {'flash inside': np.zeros(num_flashes),
                    'area total': 0,
                    'area with flash': 0}
    #
    if num_flashes == 0:
        # If there are no GLM flashes inside the Path domain,
        # then return None to indicate that there were no flashes.
        # We use this later to indicate a hit-rate of np.nan.
        # Also return the total contoured LTNG area. 
        fxx_stats[i]['flash inside'] = None
        for j, V in enumerate(COLLECTION.collections[level].get_paths()):
            # Compute area of the contours
            obj = {'type':'Polygon','coordinates':[[tuple(i) for i in V.vertices]]}
            json_area = geojson_area(obj)/1e6 # Units: square kilometers
            # Add area ot area_total and area_w_flash
            fxx_stats[i]['area total'] += json_area
    else:
        # Combine the GLM lat/lon points into a tuple pair
        latlon_pair = [(paths_glm[i]['longitude'][j], paths_glm[i]['latitude'][j]) for j in range(num_flashes)]
        #
        # Each collection may have many paths--a path for each contour area.
        # We need to loop through each area and find out which points
        # lie inside the area. Those points that inside the path area (True),
        # Then we add those to the inside array.
        num = len(COLLECTION.collections[level].get_paths())
        for j, V in enumerate(COLLECTION.collections[level].get_paths()):
            # 1) Determine which flashes are inside each contour
            is_inside = V.contains_points(latlon_pair)
            fxx_stats[i]['flash inside'] += is_inside
            #
            # 2) Was there any flash in the area?
            any_flash = np.sum(is_inside) > 0
            #
            # 3) Compute area of the contour
            obj = {'type':'Polygon','coordinates':[[tuple(i) for i in V.vertices]]}
            json_area = geojson_area(obj)/1e6 # Units: square kilometers
            #
            # Add area to area_total and area_w_flash
            fxx_stats[i]['area total'] += json_area
            if any_flash:
                fxx_stats[i]['area with flash'] += json_area
            #
            sys.stdout.write('\r-->>  %s: %.1f%% Complete (%s of %s contour areas)' % (i, (j+1)/num*100, (j+1), num))
            #
        # Convert the inside array from a count to a boolean. This boolean
        # matches the lat/lon points in the same order.
        # Add this information to the paths_glm file as a new key 'in_LTNG'
        # meaning the GLM point is inside the HRRR's LTNG field.
        fxx_stats[i]['flash inside'] = np.array(fxx_stats[i]['flash inside'], dtype=bool)
```
