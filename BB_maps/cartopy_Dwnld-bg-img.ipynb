```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
```

```python
import os
os.environ["CARTOPY_USER_BACKGROUNDS"] = "/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v3/BB_maps/BG/"
```

```python
ax = plt.axes(projection=ccrs.LambertConformal(central_longitude=-100))
ax.coastlines()
ax.background_img(name='Terrain Greyscale', resolution='high')
ax.set_extent([-120, -75, 23, 50])

ax.add_wms(wms='https://mesonet.agron.iastate.edu/cgi-bin/wms/nexrad/n0q-t.cgi?',
           layers='nexrad-n0q-wmst', zorder=10)
```
