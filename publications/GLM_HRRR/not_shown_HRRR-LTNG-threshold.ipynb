# _Not Shown_: Definition of HRRR lightning threshold.
Show the difference between a threshld of 0.00 and 0.04 flashes/km2/5-min

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import multiprocessing
import os

import sys
sys.path.append('/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v3/')
from BB_HRRR.HRRR_Pando import get_hrrr_variable, get_hrrr_latlon
from BB_maps.my_basemap import draw_HRRR_map, draw_centermap
from BB_cmap.reflectivity_cmap import reflect_ncdc

from fires_list import get_fire
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,5))

c = reflect_ncdc()
events = ['Mallard', 'Lake Christine', 'July Storm']
FXX = 6

for event, ax in zip(events, axes):
    FIRE = get_fire(event)
    DATE = FIRE['event']
    m = draw_centermap(center=(FIRE['latitude'], FIRE['longitude']), size=(3.5,3.5))
    H = get_hrrr_variable(DATE-timedelta(hours=FXX), 'LTNG:entire', fxx=FXX, verbose=False)
    Hrefc = get_hrrr_variable(DATE-timedelta(hours=FXX), 'REFC:entire', fxx=FXX, verbose=False)
    
    plt.sca(ax)
    m.contourf(Hrefc['lon'], Hrefc['lat'], Hrefc['value'], latlon=True, levels=range(0,80,5),
               cmap=c['cmap'], vmax=c['vmax'], vmin=c['vmin'], alpha=.1)
    
    m.contour(H['lon'], H['lat'], H['value'].data>0,
              levels=[0], latlon=True, colors='k')
    m.contour(H['lon'], H['lat'], H['value'].data>0.04,
              levels=[0], latlon=True, colors='tab:red')
    
    m.drawstates()

    plt.title('LTNG > 0.0 (black)\nLTNG > 0.04 (red)')
    plt.xlabel('F%02d valid: %s' % (FXX, H['valid'].strftime('%H%M UTC %d %b %Y')))
```
